```python
# Load the dataset
df = pd.read_csv("Home.csv")
```

```python
print(df.head())
```

```python
# Create an interaction term for temperature and humidity
df['tem_hum_interaction'] = df['tem'] * df['hum']
```

```python
# Remove the 'date' and 'occ' column from the dataset
df.drop('date', axis=1, inplace=True)
df.drop('occ', axis=1, inplace=True)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Check the size of the DataFrame
num_rows, num_columns = df.shape

# Print the number of rows and columns
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
from scipy.stats import zscore
import matplotlib.pyplot as plt
# Calculate z-scores for numerical variables
numeric_vars = df.select_dtypes(include='number')
z_scores = zscore(numeric_vars)


# Visualize the outliers
for col in numeric_vars.columns:
    plt.figure(figsize=(10, 4))
    
    # Box plot
    plt.subplot(1, 2, 1)
    df.boxplot([col])
    
    # Histogram
    plt.subplot(1, 2, 2)
    df[col].hist(bins=30)
    
    plt.tight_layout()
    plt.show()

# Set the z-score threshold for outlier removal
threshold = 3

# Identify rows with outliers
outliers = (z_scores > threshold) | (z_scores < -threshold)

# Remove rows with outliers and create a copy
df = df[~outliers.any(axis=1)].copy()

# Print the modified dataset without outliers
print(df)
```

```python
# ---- Start of Feature Scaling -----

# Additional import
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the numeric columns of the data
scaled_features = scaler.fit_transform(df.drop(columns=['occ_int']))

# Overwrite df_no_outliers with the scaled data, while preserving column names and indices
df.loc[:, df.columns != 'occ_int'] = scaled_features

# ----- End of Feature Scaling ------
```

```python
# Check the size of the DataFrame
num_rows, num_columns = df.shape

# Print the number of rows and columns
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
# Visualize the correlation matrix
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Visualize the relationship between pressure and temperature
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='pre', y='tem')
plt.title("Pressure vs Temperature")
plt.xlabel("Pressure")
plt.ylabel("Temperature")
plt.show()
```

```python

# Visualize the relationship between humidity and temperature
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='hum', y='tem')
plt.title("Humidity vs Temperature")
plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.show()
```

```python
# Split the dataset into features and target
features = df.drop(columns=['occ_int'])
target = df['occ_int']
```

```python
# Perform stratified k-fold cross-validation
k = 5  # Number of folds
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for train_index, test_index in skf.split(features, target):
    train_features, test_features = features.iloc[train_index], features.iloc[test_index]
    train_target, test_target = target.iloc[train_index], target.iloc[test_index]
```

```python
# Initialize TabNet classifier
learning_rate = 0.001  # Set the desired learning rate
model = TabNetClassifier(optimizer_params={"lr": learning_rate})
```

```python
# Train the model
model.fit(
    X_train=train_features.values,
    y_train=train_target.values,
    max_epochs=30,
    patience=10,
    batch_size=32,
    virtual_batch_size=16,
)
```

```python
# Access the feature importance scores
feature_importances = model.feature_importances_

# Display feature importance scores
plt.figure(figsize=(8, 6))
plt.bar(features.columns, feature_importances)
plt.title("Feature Importance")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.xticks(rotation=45)
plt.show()
```

```python
# Save the trained model
model.save_model("tabnet_model_04")
```

```python
# Load the trained model
loaded_model = TabNetClassifier()
loaded_model.load_model("tabnet_model_04.zip")
```

```python

# Make predictions on the test set
predictions = loaded_model.predict(test_features.values)
```

```python
# Evaluate the model
accuracy = (predictions == test_target.values).mean()
print("Accuracy:", accuracy)
```

```python
# Confusion Matrix
conf_mat = confusion_matrix(test_target, predictions)
plt.figure(figsize=(8, 6))
plt.imshow(conf_mat, cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks(range(len(set(target))), set(target), rotation=45)
plt.yticks(range(len(set(target))), set(target))
plt.show()

print("Confusion Matrix:")
print(conf_mat)
```

```python
# Classification Report
report = classification_report(test_target, predictions)
print("Classification Report:\n", report)
```
